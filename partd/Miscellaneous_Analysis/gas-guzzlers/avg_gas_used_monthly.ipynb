```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/avg_gas_used_year.txt', delimiter=", \[", engine='python', header=None)
df = df1.T
```

```python
df.columns = ['Name']
```

```python
new = df["Name"].str.split(",", n = 1, expand = True)
 
# making separate first name column from new data frame
df["Monthly_period"]= new[0]
 
# making separate last name column from new data frame
df["avg_gas_used_year"]= new[1]
 
# Dropping old Name columns
df.drop(columns =["Name"], inplace = True)

df = df.replace(regex=r'[\[\]]', value='')
```

```python
df.head()
```

```python
df["avg_gas_used_year"] = pd.to_numeric(df["avg_gas_used_year"].str.replace(r'\D', ''), errors='coerce').astype(float)
df.dtypes
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year')
plt.ylabel('Average gas used')
x=df['Monthly_period']
y=df['avg_gas_used_year']
plt.plot(x,y,color='blue')
```
