```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/avg_gas_price.txt', delimiter=", \[", engine='python', header=None)
df = df1.T
```

```python
df.columns = ['Name']
```

```python
new = df["Name"].str.split(",", n = 1, expand = True)
 
# making separate first name column from new data frame
df["Monthly_period"]= new[0]
 
# making separate last name column from new data frame
df["avg_gas_price"]= new[1]
 
# Dropping old Name columns
df.drop(columns =["Name"], inplace = True)

df = df.replace(regex=r'[\[\]]', value='')
```

```python
df.head()
```

```python
df["avg_gas_price"] = df[["avg_gas_price"]].apply(pd.to_numeric)
df.dtypes
```

```python
plt.figure(figsize = (30,10))
plt.xticks(rotation=90)
plt.xlabel('period')
plt.ylabel('avg_gas_price')
plt.bar(df['Monthly_period'],  df['avg_gas_price'])
```
